```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Defining video path / output directory  
video_path = os.path.join(os.path.expanduser('~'), "Desktop", "IMG_3487.MOV")
output_path = os.path.join(os.path.expanduser('~'), "Desktop", "output")
```

```python
video = cv2.VideoCapture(video_path)
```

```python
# FPS
fps = video.get(cv2.CAP_PROP_FPS) 
```

```python
# FRAME COUNT
frame_count = video.get(cv2.CAP_PROP_FRAME_COUNT)
```

```python
# Setting video to end
video.set(cv2.CAP_PROP_POS_FRAMES, frame_count -1)
video.get(cv2.CAP_PROP_POS_FRAMES)
```

```python
# DURATION
duration = frame_count/fps * 1000
```

```python
# Setting video to end
max_msec = duration - 1/fps * 1000
video.set(cv2.CAP_PROP_POS_MSEC, max_msec)
```

```python
# Showing frame
ret, frame = video.read()
plt.imshow(frame)
```

```python
msec = np.linspace(0, max_msec, 10)
```

```python
for idx, time in enumerate(msec): 
    video.set(cv2.CAP_PROP_POS_MSEC, time)
    ret, frame = video.read()
    photo_capture = frame[900:1100, 500:600]
    cv2.imwrite(os.path.join(output_path, f'frame_{idx}.png'), photo_capture)
```
